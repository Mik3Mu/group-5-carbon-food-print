```python
# The Score is devided in 5 categories : A , B , C , D , E
# A is True if "CO2/kg" <= 1
# B is True if "CO2/kg" >= {3,1}
# C is True if "CO2/kg" >= {5,3}
# D is True if "CO2/kg" >= {10,5}
# E is True if "CO2/kg" >  10

def scoring(CO2):
    if CO2 <= 1:
        score = "A"
    elif CO2 <= 3:
        score = "B"   
    elif CO2 <= 5:
        score = "C"
    elif CO2 <= 10:
        score = "D"
    else: 
        score = "E"
    return score
```

```python
df["Score"] = [
    scoring(CO2)
    for CO2 in df.CO2_Equivalent_per_Kilo
    ]
df
```
