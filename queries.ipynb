```python
import pandas as pd
```

```python
F_product = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "F_product")
F_category = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "F_category")
D_product_category = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_category")
D_product_processing = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_proces")
D_product_animal_feed = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_animal")
D_product_land_use = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_land_u")
D_product_farm = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_farm")
D_product_transport = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_transp")
D_product_retail = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_retail")
D_product_packaging = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_packag")
D_product_CO2_equivalent = pd.read_excel("Data base/Normalized_data.xlsx",sheet_name = "D_product_food_emissions_CO2_eq")
```

```python
M_product_m = pd.merge(D_product_category,D_product_processing,how="left",left_on="product",right_on="product")
M_product_m1  = pd.merge(M_product_m,D_product_animal_feed,how ="left",left_on="product",right_on="product")
M_product_m2 = pd.merge(M_product_m1,D_product_land_use,how ="left",left_on="product",right_on="product")
M_product_m3 = pd.merge(M_product_m2,D_product_farm,how ="left",left_on="product",right_on="product")
M_product_m4 = pd.merge(M_product_m3,D_product_transport,how ="left",left_on="product",right_on="product")
M_product_m5 = pd.merge(M_product_m4,D_product_retail,how ="left",left_on="product",right_on="product")
M_product_m6 = pd.merge(M_product_m5,D_product_packaging,how ="left",left_on="product",right_on="product")
M_product_list = pd.merge(M_product_m6,D_product_CO2_equivalent,how ="left",left_on="product",right_on="product")
M_product_list
```

```python
M_product_list.to_csv("Normalized_data_cleared", index =False)
```
